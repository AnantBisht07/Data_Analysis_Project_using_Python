```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as  px
```

```python
dt = pd.read_csv('apple_products.csv')
dt
```

### First let’s have a quick look at whether this dataset contains any null values or not

```python
dt.isnull().sum()
```

### Now, let’s have a look at the descriptive statistics of the data with the describe method

```python
dt.describe()
```

###  Now i will create a new dataframe that is of top 15 rated iphones and it helps to understand what kind of iPhones are liked the most 

```python
top_rated = dt.sort_values(by=['Star Rating'],
                            ascending = False)
top_rated = top_rated.head(15)
print(top_rated['Product Name'])
```

### here are the top rated iphones list

```python
 top_rated['Product Name']
```

### Now let’s have a look at the number of ratings of the highest-rated iPhones

```python
iphones = top_rated["Product Name"].value_counts()
label = iphones.index
counts = top_rated["Number Of Ratings"]
figure = px.bar(top_rated, x = label, 
                y = counts, 
            title="Number of Ratings of Highest Rated iPhones")
figure.show()
```

### Now let’s have a look at the number of reviews of the highest-rated iPhones 

```python
iphones = top_rated["Product Name"].value_counts()
lablel = iphones.index
counts = top_rated["Number Of Reviews"]
figure = px.bar(top_rated, x = label, 
                y = counts, 
            title="Number of Reviews of Highest Rated iPhones")
figure.show()
```

### Now let’s have a look at the relationship between the sale price of iPhones and their ratings
#### There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more in India

```python
figure = px.scatter(data_frame = dt,x="Number Of Ratings",
                    y="Sale Price", size= "Discount Percentage", 
                    trendline= "ols", 
                    title= "Relationshiop Between SALE PRICE AND NUMBER OF RATINGS")
figure.show()
```

### Now let’s have a look at the relationship between the discount percentage on iPhones and the number of ratings
#### There is a positive linear relationship between the discount percentage on iPhones  and the number of ratings. It means iPhones with high discounts are sold more in India.

```python
figure = px.scatter(data_frame = dt, x= "Number Of Ratings",
                   y= "Discount Percentage",
                   size = "Sale Price",
                   trendline= "ols",
                   title= "Relationshiop Between Discount Percentage AND NUMBER OF RATINGS")
figure.show()
```

### Now we see the relationship between sale price of iphone and star rating
####  There is a positive linear relationship between the sale price on iPhones and thestar rating. It means iPhones with less sale price  are having more star rating in India.

```python
figure = px.scatter(data_frame= dt, x= "Sale Price",
                   y= "Star Rating",
                   size= "Number Of Reviews",
                   trendline= "ols",
                   title= "Relationship Between Sale Price AND Star Rating")
figure.show()
```

### Comparison of Product Name and Mrp

```python
top_10 = dt.sort_values(by='Mrp', ascending=False).head(10)
plt.bar(top_10['Product Name'], top_10['Mrp'])
plt.xlabel('Product Name')
plt.ylabel('Mrp')
plt.title('Comparison of Product Name and Mrp')
plt.xticks(rotation=90)
plt.show()
```
